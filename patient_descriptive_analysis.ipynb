```python
#!pip install tableone
import pandas as pd
from tableone import TableOne
import warnings

#import ace_tools as tools
warnings.filterwarnings("ignore")
pd.set_option('display.max_rows', 200)  # Set to 200 or more, higher than 105
pd.set_option('display.max_columns', None)  # Show all columns if needed
```

```python
patient_level_df = pd.read_csv('alert_analysis/data/main_data_2022/df_patients_level_data.csv')
print(patient_level_df.shape)
patient_level_df.head(10)
```

```python
patient_level_df.columns.to_list()
```

```python
# Define categorical and continuous variables
categorical_vars = [
    "age_category",
    'unit_Internal',
    'unit_Internal-Covid19',
     'unit_Surgery',
     'unit_Gynecology',
     'unit_Cardiology',
     'unit_Emergency',
     'unit_Geriatric',
     'unit_Hematology',
     'unit_Nephrology',
     'unit_Oncology'
]

continuous_vars = [
    "age",
    'survival_rate_10y_age_adj_mean',
    'charlson_score_age_adj_mean',
    'hospital_days_mean',
    'chronic_med_count_mean',
    'diagnosis_count_mean'
    
]

#non_normal = ["age"]

# Generate summary table
table = TableOne(
    patient_level_df,
    categorical=categorical_vars,
    groupby="gender",
    continuous=continuous_vars,
    #nonnormal=non_normal,
    pval=True,  # Add p-values
    missing=True  # Show missing values
)

# Print the table
print(table.tabulate(tablefmt="pipe"))
```

```python
continuous_vars = [
    "age",
]

non_normal = ["age"]

# Generate summary table
table = TableOne(
    patient_level_df,
    #categorical=categorical_vars,
    groupby="gender",
    continuous=continuous_vars,
    nonnormal=non_normal,
    pval=True,  # Add p-values
    missing=True  # Show missing values
)

# Print the table
print(table.tabulate(tablefmt="pipe"))
```
